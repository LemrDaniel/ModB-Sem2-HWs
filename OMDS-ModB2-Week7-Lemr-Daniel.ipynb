# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

In order to make people focus to be on the last three months of data first, I would suggest increasing the line thickness and darkening the color of the line in that area. That would draw the eye to that part of the page first. Since the right side seems to be at the extremems of the overall increase or decrease, it might make sense to add data labels to just the last three point of each graph as well. That would help with retention on the last three points.

```python
#Import 3rd Dataset
#Prep work from HWs

delay_causes = pd.read_csv(r"C:\Users\lemrd\Downloads\OMDS\Mod B\Sem 2\Data\Flight_Delay_and_Causes_Data\Flight_delay.csv")

delay_causes.drop(["Cancelled","Diverted","CancellationCode"],axis=1, inplace=True)
delay_causes.drop_duplicates(keep="first", inplace=True)
                           
delay_causes.head()
```

```python
delay_causes.plot.scatter("CRSArrTime", "ArrTime")

plt.scatter(delay_causes["CRSArrTime"], delay_causes["ArrTime"], label = "Corr = .50")
plt.xlabel("Expected Arrival Time")
plt.ylabel("Actual Arrival Time")
plt.title("Raw Data Arrival Times (Low Correlation)")
plt.savefig("Raw_Arr_Diff.png")
```

```python
data_to_shift = delay_causes[delay_causes["ArrTime"] < (delay_causes["CRSArrTime"] - 100)].index
delay_causes.loc[data_to_shift,"ArrTime"] = delay_causes.loc[data_to_shift,"ArrTime"] + 2400

plt.scatter(delay_causes["CRSArrTime"], delay_causes["ArrTime"], label = "Corr = .98")
plt.xlabel("Expected Arrival Time")
plt.ylabel("Actual Arrival Time")
plt.title("Modified Data Arrival Times (High Correlation)")
plt.savefig("Mod_Arr_Diff.png")
```
